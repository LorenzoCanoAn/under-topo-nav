```python
import pickle
import torch
import os
import matplotlib.pyplot as plt
ORIGIN_FOLDER = "/home/lorenzo/Datasets/gallery_detection/laserscan_image_polished"
DESTINATION_FOLDER = "/home/lorenzo/catkin_data/datasets/2d_gallery_detection/transformed_old_dataset"
```

```python
for i, fnw in enumerate(os.listdir(ORIGIN_FOLDER)):
    fpr = os.path.join(ORIGIN_FOLDER,fnw)
    fpw = os.path.join(DESTINATION_FOLDER,  f"{i}.pickle")
    with open(fpr, "rb") as fr:
        image, label = torch.load(fr)
        image /= torch.max(image)
        with open(fpw, "wb+") as fw:
            pickle.dump((image,label),fw)
    print(i,end="\r")
        
```

```python
for i, fn in enumerate(os.listdir(DESTINATION_FOLDER)):
    fpr = os.path.join(DESTINATION_FOLDER,fn)
    with open(fpr, "rb") as fr:
        image, label = pickle.load(fr)
        plt.imshow(image)
        plt.show()
        plt.plot(label)
        plt.show()
    if i > 20:
        break
        
```
